Jadi pada kolom NEIGHBORHOOD terdapat 254 nilai unik (NaN tidak terhitung), BUILDING CLASS CATEGORY memiliki 47 nilai unik dan seterusnya. Semua feature yang disebutkan diatas merupakan feature yang bernilai kategorik.

```python
# mengkonversi tipedata feature 'LAND SQUARE FEET', 'GROSS SQUARE FEET', 'SALE PRICE', 'BOROUGH' menjadi numerik
numer= ['LAND SQUARE FEET', 'GROSS SQUARE FEET', 'SALE PRICE', 'BOROUGH']
for col in numer:
    data[col]= pd.to_numeric(data[col], errors='coerce')
    
# mengkonversi tipedata feature 'NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'TAX CLASS AT PRESENT', 'BUILDING CLASS AT TIME OF SALE', 'TAX CLASS AT TIME OF SALE', 'BOROUGH' menjadi kategorik
categ= ['NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'TAX CLASS AT PRESENT', 'BUILDING CLASS AT TIME OF SALE', 'TAX CLASS AT TIME OF SALE', 'BOROUGH']
for col in categ:
    data[col]= data[col]. astype('category')
    
# mengkonversi tipedata 'SALE DATE'
data['SALE DATE']= pd.to_datetime(data['SALE DATE'], errors='coerce')
```

## Analyze Confidence Intervals

Beapakah interval kepercayaan bahwa proporsi rata-rata harga jual dapat mewakili harga jual sebenarnya ?

```python
data.groupby('BOROUGH').agg({"SALE PRICE": [np.mean, np.std, np.size]})
```

```python
mean_mht = 2919586 # rata-rata harga properti di Manhatan
sd_mht = 21527570 # standar deviasi harga properti di Manhattan
n = 18102 # jumlah properti di Manhattan
z = 1.96 # z-score berdasarkan tabel z
se = sd_mht / np.sqrt(n) # menghitung standart eror

# menghitung interval kepercayaan
lower = mean_mht - z*se
upper = mean_mht + z*se
print("(",lower,",",upper,")")
```

Berdasarkan hasil tersebut diperoleh bahwa rata-rata harga properti di Manhattan adalah antara 2605977.5173839475 dan 3233194.4826160525

## Hypothesis Testing

Dengan mempertimbangkan data NYC Property, apakah Queens memiliki harga jual rata-rata yang lebih tinggi secara signifikan daripada Broklyn ?

Population : sale price in the NYC Property data

Parameter of interest: \mu1-\mu2, Sale Price

H_0: rata-rata harga jual properti di Queens adalah sama dengan rata-rata harga jual di Queens

H_1: rata-rata harga jual properti di Brooklyn tidak sama dengan rata-rata harga jual di Queens

Data :

26548 properti di Queens

\mu1= 904971,8

\sigma1= 2699175


23843 properti di Broklyn

\mu2= 841083,4

\sigma2= 3951912

\mu1-mu2= 63888,4

```python
queens = data[data["BOROUGH"] == 4]
brooklyn = data[data["BOROUGH"] == 3]
n_queens = len(queens)
mu_queens = queens["SALE PRICE"].mean()
sd_queens = queens["SALE PRICE"].std()
(n_queens, mu_queens, sd_queens)
```

```python
n_br = len(brooklyn)
mu_br = brooklyn["SALE PRICE"].mean()
sd_br = brooklyn["SALE PRICE"].std()
(n_br, mu_br, sd_br)
```

```python
sm.stats.ztest(queens["SALE PRICE"].dropna(),
               brooklyn["SALE PRICE"].dropna(),
              alternative='two-sided')
```

Karena p-value= 0,0326 < 0.05 maka keputusannya adalah tolak H_0 dan terima H_1
sehingga dapat disimpulkan bahwa rata-rata harga jual properti di Brooklyn tidak sama dengan rata-rata harga jual di Queens
